```python
import matplotlib.pyplot as plt
from scipy.interpolate import approximate_taylor_polynomial
import numpy as np
```

```python
f = lambda x: 1/((x**2) + 2)
# Definir función a aproximar

a = 0
# Definir variable de punto sobre el cual la función será centrada

grados = range(1, 6)
# Grados de polinomios sobre los cuales serán definidos los polinomios de Taylor
```

```python
x = np.linspace(-10.0, 10.0, num=100)
# Valores a graficar eligidos arbitrariamente

plt.figure(figsize=(15, 10))
# Definir figura sobre la cual existe la gráfica

plt.plot(x, f(x), label="Función", linewidth=3)
#Graficar función definida en los puntos elegidos arbitrariamente

errores_truncamiento = []
# Iniciar lista sobre la cual será insertado cada error de truncamiento

for grado in grados:
    # Por cada grado

    taylor = approximate_taylor_polynomial(f, a, grado, 1,
                                               order=grado)
    
    # Definir variable "taylor" igual a la función polinomios de Taylor con grado seleccionado del rango especificado, (Uno, Dos, Tres, Cuatro y por último Cinco grados)
    # el cual luego espera parámetro x que es el número a sustituir en la fórmula de los polinomios.

    plt.plot(x, taylor(x), label=f"Grado={grado}", linewidth=1.2)
    # Graficar polinomio de Taylor con el grado correspondiente al elemento seleccionado en la iteración

    errores_truncamiento.append(f(x) - taylor(x))
    # Insertar error de truncamiento el cual es dado por el valor de la función en x y su aproximación utilizando la función definida como "taylor"

plt.legend()
plt.tight_layout()

plt.axis([-15, 15, -15, 15])
# Especificar ejes para mejor visualización

plt.show()
```

```python
def graficar_errores_truncamiento(errores):
    f = plt.figure(figsize=(17, 15))
    for i in range(len(errores)):
        ax = plt.subplot(3, 3, i+1)
        plt.plot(errores[i], linewidth=3)
        plt.title(f"Error Aproximación de Polinomios grado {i+1}")
    


graficar_errores_truncamiento(errores_truncamiento)
```

## Teoría



Rodolfo Louro define los polinomios de Taylor de la siguiente manera: **"**Sea una función $f : I → R$ ($I$ un intervalo) tal que $f$ admite derivadas de orden $n$ en cierto entorno $E(a)$, con $a ∈ I$  . Llamaremos polinomio de Taylor (Brook Taylor (1685-1731) ) de orden $n$ generado por $f$ en $x = a$  al siguiente polinomio: $P_n(x)=f(a) + \frac{f'(a)}{!1}(x-a) + \frac{f''(a)}{!2}(x-a)^2 + ... + \frac{f^{n}(a)}{!n}(x-a)^n$ **"**

Añade el Prof. Jorge Brisset que **"**En los casos que trabajamos con $a = 0$ , los polinomios reciben
el nombre de polinomios de Mac Laurin (Colin Mac Laurin (1698-1746) )**"**

## Análisis

De los resultados obtenidos en cada polinomio se observa que, los errores en su mayoría son casi nulos ($error \approx 0$) y se prolongan cada vez más conforme el número del polinomio seleccionado aumenta
